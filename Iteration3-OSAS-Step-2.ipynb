## 2.3 Explore the Data

### Loan Status

```python
import matplotlib.pyplot as plt
vdf = df[df.loan_status != 'Does not meet the credit policy. Status:Fully Paid']
vdf = vdf[vdf.loan_status != 'Does not meet the credit policy. Status:Charged Off']
vdf.loan_status.value_counts().sort_values().plot(kind='barh')
plt.title('Frequency of Loan Status', size = 15)
plt.ylabel('Loan Status', fontsize = 13)
plt.xlabel('Frequency', fontsize = 13)
plt.show()
```

```python
vdf = df[df.loan_status != 'Does not meet the credit policy. Status:Fully Paid']
vdf = vdf[vdf.loan_status != 'Does not meet the credit policy. Status:Charged Off']
labels = list(vdf.loan_status.value_counts().index)
values = list(vdf.loan_status.value_counts().values)
fig, pos = plt.subplots(figsize=(8,8))
explode = (0.1,0.1,0,0,0,0,0,0)
pos.pie(values, labels=labels, explode=explode, autopct='%1.1f%%',shadow=True, startangle=90)
pos.axis('equal')
pos.set_title('Proportion of Loan Status', size = 15)
plt.show()
```

```python
vdf = df[df.loan_status != 'Does not meet the credit policy. Status:Fully Paid']
vdf = vdf[vdf.loan_status != 'Does not meet the credit policy. Status:Charged Off']
vdf = vdf.groupby('loan_status')['loan_amnt'].mean()
vdf = vdf.reset_index().sort_values('loan_amnt')
labels = vdf.loan_status.values
values = vdf.loan_amnt.values

fig, pos = plt.subplots(figsize=(10,5)) 
ind = np.arange(len(values))
width = 0
pos.barh(ind, values)
pos.set_yticks(ind+width/2)
pos.set_yticklabels(labels, minor=False)
pos.set_title('Average Loan Amount by Loan Status', size = 15)
pos.set_xlabel('Average Loan Amount', fontsize = 13)
pos.set_ylabel('Loan Status', fontsize = 13)
for i, j in enumerate(values):
    pos.text(j + 900, i + .1, str(round(j)), color='blue')
plt.show()
```

## 2.4 Verify the Data Quality

```python
dfsize = len(df)
print('{:<30} {:^6}'.format('Attributes','No. of Missing Values'))
for i in df:
    if df[i].isnull().sum() != 0:
        missing = dfsize - df[i].isnull().sum()
        print('{:<30} {:^6}'.format(i, str(missing)))
```

### Outliers

```python
outliers ={}
print('{:<25} {:^6}'.format('Attributes','No. of Outliers'))
for i in df:
    if df[i].dtype != 'object':
        q3, q1 = np.percentile(df[i].dropna(), [75,25])
        upper = q3 + (q3 - q1) * 1.5 
        lower = q1 - (q3 - q1) * 1.5 
        a = len(df.iloc[np.where(df[i].dropna() > upper)]) + len(df.iloc[np.where(df[i].dropna() < lower)])
        if a != 0:
            print('{:<30} {:^6}'.format(i,a))
            outliers[i] = a
outliers = pd.DataFrame([outliers.keys(), outliers.values()]).T
```
